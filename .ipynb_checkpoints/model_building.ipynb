```python
# Import Statements

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Load datasets

import os

file_path = "../stock_data_final.csv"
if not os.path.exists(file_path):
	raise FileNotFoundError(
		f"{file_path} not found. "
		f"Make sure the CSV is located in the working directory: {os.getcwd()}"
	)

df = pd.read_csv(file_path)

# Drop rows with missing values (if any)
df = df.dropna()

df.head()
```

```python
X = df.drop(columns=["Close"])
y = df["Close"]
if "Date" in X.columns:
    X = X.drop(columns=["Date"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

y_pred_lr = lr_model.predict(X_test)
```

```python
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
```

```python
def evaluate_model(true, pred, model_name):
    mae = mean_absolute_error(true, pred)
    rmse = np.sqrt(mean_squared_error(true, pred))
    r2 = r2_score(true, pred)

    print(f"\n===== {model_name} =====")
    print("MAE:", mae)
    print("RMSE:", rmse)
    print("R2 Score:", r2)

    return mae, rmse, r2
```

```python
lr_metrics = evaluate_model(y_test, y_pred_lr, "Linear Regression")
rf_metrics = evaluate_model(y_test, y_pred_rf, "Random Forest")
```

```python
results = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest"],
    "MAE": [lr_metrics[0], rf_metrics[0]],
    "RMSE": [lr_metrics[1], rf_metrics[1]],
    "R2 Score": [lr_metrics[2], rf_metrics[2]]
})

results
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test.values, label="Actual", color="blue")
plt.plot(y_pred_rf, label="Predicted (RF)", color="red")
plt.legend()
plt.title("Actual vs Predicted - Random Forest")
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.show()
```
