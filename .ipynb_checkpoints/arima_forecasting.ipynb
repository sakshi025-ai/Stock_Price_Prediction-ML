```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
df = pd.read_csv("../stock_data_final.csv")

df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")
df.set_index("Date", inplace=True)

close_series = df["Close"]
```

```python
model = ARIMA(close_series, order=(5,1,0))
arima_model = model.fit()

print(arima_model.summary())
```

```python
forecast = arima_model.forecast(steps=30)
```

```python
plt.figure(figsize=(12,6))
plt.plot(close_series, label="Historical Data")
plt.plot(forecast, label="30-Day Forecast", color="red")
plt.legend()
plt.title("ARIMA 30-Day Stock Price Forecast")
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.show()
```
